```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
Faker.seed(9999)
contacts = pd.read_csv('contacts.csv')
header = ["First", "Last", "Age", "Phone", "E-mail", "Address"]
```

```python
first = [fake.first_name() for i in range(5000)]
last = [fake.last_name() for i in range(5000)]
age = [random.randint(18, 100) for i in range(5000)]
phone = [fake.phone_number() for i in range(5000)]
email = [fake.email() for i in range(5000)]
address = [fake.address() for i in range(5000)]
# contact_info = pd.DataFrame(data=[first, last, age, phone, email, address], index=[0], columns=header)
contact_info = pd.DataFrame(data={'First': first,
                                 'Last' : last,
                                 'Age': age,
                                 'Phone': phone,
                                 'E-mail': email,
                                 'Address': address}, columns=header)
```

```python
contact_info.to_csv('contacts.csv',mode='a',index=False, header=False)
```
